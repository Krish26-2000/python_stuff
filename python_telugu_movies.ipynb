```python
from googlesearch import search as goog_search
import numpy as np
import sqlite3
```

```python
years = np.arange(2011, 2021)
dbase = "telugu_cinema.db"
db_conn = sqlite3.connect(dbase)
for year in years:
    db_conn.execute(f"DROP TABLE IF EXISTS {year}")
    db_conn.execute(f"""CREATE TABLE {year} (
                   Name VARCHAR(255) NOT NULL,
                   link VARCHAR NOT NULL,
                   Budget_crores FLOAT,
                   Box_office_crores FLOAT
                   )""")
    
    text2search = f"telugu movies released in {year} site:wikipedia.org"
    for link in goog_search(text2search, num = 1, stop=1):
        print(link)
```
